```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
from PIL import Image
```

```python
model = YOLO('yolov8n.pt')

results = model.train(data='smoking.yaml', epochs=5)
```

```python
def classify_image(image_path):
    results = model(image_path)

    for result in results:
        for box in result.boxes:
            if box.cls[0] == 0:
                return "Smoking"

    return "No Smoking"

def display_image_with_prediction(image_path, results, prediction):
    img = Image.open(image_path)
    plt.imshow(img)
    plt.axis('off')
    plt.title(prediction)

    for result in results:
        for box in result.boxes:
            x1, y1, x2, y2 = box.xyxy[0].tolist()
            plt.gca().add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1, fill=False, edgecolor='red', linewidth=2))
    plt.show()

image_path = '05-12-21-happy-people.jpeg'
results = model(image_path)
prediction = classify_image(image_path)
display_image_with_prediction(image_path, results, prediction)
```
